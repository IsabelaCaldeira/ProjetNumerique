## Partie I. Fonctions

Les fonctions sont particulièrement utiles pour effectuer des opérations amenées à être répétées un grand nombre de fois. Au lieu de réécrire les mêmes lignes tout au long d'un code, il suffit de définir une fois la fonction et l'appeler autant de fois que nécessaire.
En python, pour définir une fonction, on utilise la syntaxe $\mathtt{def}\, \mathtt{Nom}(\mathtt{arguments}):$ 
    
où "$\mathtt{arguments}$" désigne le ou les paramètres nécessaires pour définir la fonction (il est aussi possible d'avoir une liste vide). La fonction donnera - ou pas - en sortie un ou des résultats.
                                                                                
Au sein d'une fonction, l'**indentation** est obligatoire (touche "Tab" du clavier). Pour sortir de la fonction, il suffit donc de revenir à la ligne en retirant l'indentation. 
                                                                                

Par exemple : 

```python
def MaFonction(x):
    return(2*x)
print(MaFonction(3))
MaFonction(7)
```

La syntaxe $\mathtt{def}$ permet donc de définir le nom de la fonction (ici "$\mathtt{MaFonction}$"). 
Après le nom, les parenthèses contiennent le ou les arguments de la fonction séparés les uns des autres par une virgule. Dans l'exemple, il n'y a qu'un argument "$x$". On notera que le nom (quelconque) de la variable passée ici en argument est une variable **locale** à la fonction.
La sortie est définies par $\mathtt{return}$ (pour avoir plusieurs sorties, il suffit de les séparer les unes des autres par une virgule.

Un autre exemple:

```python
def F(x1,x2):
    return(x1+x2, x1-x2 )
print(F(5,4))
```

#### 1) Définir la fonction vitesse qui donnera la vitesse atteinte par un objet chutant d'une hauteur h, puis calculer les vitesses atteintes (en km h-1) pour h = 5 m et pour h = 10 m.
Indication: la vitesse est donnée par $\sqrt{2 g h}$ où $\mathtt{g}$ est l'accélération de la pesanteur

```python
import math

def vitesse(h):
    return math.sqrt(2*9.8*h)

print(vitesse(5))
print(vitesse(10))
```

Dans les exemples précédents les fonctions avaient au moins une entrée et une sortie. Cependant ce n'est pas toujours le cas. On peut ne donner aucun argument à une fonction ou ne lui faire rien renvoyer (ne pas écrire de $\mathtt{return}$ ou écrire $\mathtt{return}()$). 

Par exemple : 

```python
def Hello():
    print("Hello world")

    
Hello()
```

La fonction $\mathtt{Hello}$ ne prend aucun argument en entrée et ne renvoie rien (pas de return). Cependant elle effectue bien une action, en l'occurence afficher une chaine de caractères. 

Quand vous écrivez une fonction en Python, il n'y a pas besoin de donner le type des arguments attendus.
Par exemple testez $\mathtt{MaFonction}$("Coucou "). Le résultat vous semble-t-il cohérent ? (il devrait l'être...)

```python
MaFonction("Coucou ")

#Le result est cohérent parce que il nous retourne 2 fois la valeur de x, dans ce cas "Coucou "
```

#### 5) A faire chez soi: Ecrire une fonction qui prend en entrée un entier $N$ et un réel positif $s$ et qui renvoie la somme des entiers à la puissance $(-s)$ de $1$ à $N$ (chaque entier $i$ est mis à la puissance $i^{-s}$ avant d'être sommé). 
#### Testez votre fonction en prenant un $N$ grand et $s=2$ et comparer la valeur obtenue à $\dfrac{\pi^2}{6}$.


*NB* : la limite de cette fonction en $N \to +\infty$ est la fonction zêta de Riemann $\zeta(s)$

```python
import numpy as np
def puissance(N,s):
    t=0
    for i in range(1,N+1):
        i=i**(-s)
        t=t+i
    return t

print("N=50 et s=2 :",puissance(50,2))
print("(pi**2)/6 =",(np.pi**2)/6)
print("Les deux valeurs précédebtes sont très proches")
```

Il est crucial de savoir lire et écrire des fichiers de données avec l'outil informatique que vous utilisez.
```python
import numpy as np
```

La fonction qui nous permettra de lire un fichier est : $\mathtt{numpy.loadtxt("nom\_fichier.txt",delimiter="delim")}$.

Cette fonction prend en entrée le nom du fichier à lire (nom complet donc avec l'extension $\mathtt{.txt}$) ainsi qu'une chaîne de caractères $\mathtt{delim}$ qui correspond au(x) caractère(s) qui sépare(nt) vos données. Généralement, ceci correspond à une virgule $\mathtt{delimiter=","}$ (pour les fichiers respectant le format $comma-separated\, values, \; csv$) ou un espace $\mathtt{delimiter=" \, "}$ selon les données contenues dans le fichier.

Cette fonction renvoie donc un tableau NumPy, $a \, priori$ à deux dimensions, chaque ligne du fichier correspondant à une ligne du tableau et chaque élément séparé par le délimiteur choisi correspondant à une colonne de la ligne considérée.

De manière plus générale, la fonction $\mathtt{open("nom\_fichier.txt","w")}$ vous permet d'ouvrir un fichier avec le nom précisé et avec les droits d'écriture. Si aucun fichier de ce nom n'existe, la fonction en crée un, s'il en existe un la fonction réécrit dessus (ce qui était là avant est effacé). Pour réaliser des caractères typographiques comme un saut de ligne il faut donc écrire une chaîne de caractère particulière qui sera interprétée par la fonction d'écriture. Ainsi pour faire un saut de ligne il faut écrire la chaîne de caractère "\n".

Une autre chose à avoir en tête quand on manipule des chaînes de caractères est la concaténation. On peut concaténer des chaînes de caractères entre elles, ce qui correspond à les mettre les unes à la suite des autres. Cette opération s'écrit simplement avec l'opérateur $\mathtt{+}$. Par exemple :

## Partie III. Routines statistiques de NumPy

La librairie NumPy permet de faire des opérations directement sur les tableaux ("array" de NumPy). Notamment les calculs statistiques usuels tels que moyenne ou variance sont accessibles via des fonctions incluses dans la librairie.

Par exemple la fonction $\mathtt{numpy.mean}$ prend en entrée un tableau NumPy et renvoie la moyenne sur ses éléments.

#### 1) Créer un tableau NumPy contenant les chiffres 3, 4 et 8 et utiliser la fonction $\mathtt{numpy.mean}$ pour en calculer la moyenne (vous pouvez vérifier par vous même que le résultat est correct).

```python
tableau =np.array([3,4,8])
print("moyenne =",np.mean(tableau))
```

#### 2) Pour le même tableau, calculer la variance et l'écart-type au moyen des fonctions $\mathtt{np.var}$ et $\mathtt{np.std}$. On rappelle que la variance est le carré de l'écart-type $V=\sigma^2$

```python
print("variance =",np.var(tableau))
print("écart-type =",np.std(tableau))
```

De nombreuses autres routines (ou fonctions) sont implémentées dans la librairie NumPy (minimum, maximum, médiane, quartiles, déciles, moyenne pondérée...). Pour plus d'informations, vous pouvez vous référer à la page : https://numpy.org/doc/stable/reference/routines.statistics.html

On notera que dans le cas de tableaux multidimensionnels, les opérations statistiques sont ambigües $a\, priori$ (comment comparer des éléments des différentes dimensions du tableau ?). Sans spécification supplémentaire, lors d'un calcul statistique, un tableau multidimensionnel sera traité comme une liste "plate" d'éléments : tous les éléments du tableau seront traités sur un pied d'égalité.

#### 3) A faire chez soi. Calculer la moyenne du tableau contenu dans la variable $\mathtt{fich}$ utilisée plus haut et vérifier que cette valeur correspond à la moyenne des moyennes des lignes de $\mathtt{fich}$ prises individuellement. Pour vérifier cela, on pourra, par exemple, écrire une boucle $\mathtt{for}$ pour éviter de devoir écrire le calcul de la moyenne de chaque ligne.

```python
print("moyenne =",np.mean(fich))

a=0
for i in fich:
    a+=(i[0]+i[1]+i[2])/3

print("moyenne des lignes =",a/10)
```

On remarque une petite différence entre les deux due à une erreur d'arrondi. Quand on utilise l'outil numérique, il faut toujours se rappeler que les calculs effectués ont une précision finie (très grande mais pas infinie). Dans certaines situations, ces erreurs numériques peuvent poser problème et doivent être gérées spécifiquement.

#### 4) A faire chez soi. Calculer la variance de $\mathtt{fich}$. Peut-on procéder de la même façon qu'à la question précédente et combiner simplement les variances des lignes prises séparément? 

```python
print("variance =",np.var(fich))

b=0
for i in fich:
    b+=np.var(i)
print("variance des lignes =",b)
```

#### 2) Pour s'amuser: Conjecture de Collatz (dite conjecture de Syracuse)
Une suite de Syracuse est une suite d'entiers naturels définie de la manière suivante : on part d'un nombre entier strictement positif ; s’il est pair, on le divise par 2 ; s’il est impair, on le multiplie par 3 et l'on ajoute 1. 
En répétant l’opération, on obtient une suite d'entiers strictement positifs. Après un certain nombre d'itérations, on obtient indéfiniment la séquence des chiffres 1, 4, 2, 1, ...

Attention: on arrêtera le calcul dès que le nombre 1 est atteint. On représentera l'évolution du nombre "$U_n$" en fonction de n (n étant le numéro d'opération).
    

```python
def syracuse(n):
    i=n
    compteur=0
    while i!=1:
        compteur+=1
        if i%2==0:
            print(i)
            i=i//2
        else:
            print(i)
            i=(i*3)+1
    print("nombre d'itération =",compteur)
    return i

syracuse(27)
```

#### 3) Oscillations d'un ressort
On considère un ressort horizontal (axe $x$) dont une extrémité est fixée à un mur vertical. L'autre extrémité, à laquelle a été fixée une petite masse $m$, a été tirée d'une distance $a$ par rapport à la position à vide, définie comme $x$=0. On considère que la masse n'est soumise suivant x qu'à la force de rappel du ressort et à une force de frottement visqueux, proportionnelle à la vitesse.

Dans ces conditions, une fois la masse lâchée depuis la position $x=a$, la masse va osciller autour de la position $x$ = 0 pendant un certain temps.
L'équation horaire s'écrit:
    
##### $x \ (t) = a \ exp( - \gamma \ t /2 ) \ cos \ (wt)$
    
où $\gamma$ et $w$ sont des constantes qui dépendent des caractéristiques du système.

On considère $a$ =3 cm, $\gamma$ = 0.13 SI, $w$ = 0.9 SI. 
Préciser les unités de $\gamma$ et $w$.
Tracer le graphe de $x \ (t)$ pour $t$ variant de 0 à 100 s.
On ajoutera les lignes d'enveloppe de $x \ (t)$.
    

```python
a=0.03 #m
gamma=0.13 #kg/s/m
omega=0.9 #rad/s
t=np.linspace(0,100,1000)
x_t=a*np.exp(-gamma*t/2)*np.cos(omega*t)
enveloppe_haut=a*np.exp(-gamma*t/2)
enveloppe_bas=-a*np.exp(-gamma*t/2)

plt.plot(t,x_t,"-r")
plt.plot(t,enveloppe_haut,"--b")
plt.plot(t,enveloppe_bas,"--b")
plt.title("Oscillations avec enveloppes")
plt.xlabel("Temps (s)") ; plt.ylabel("Position (m)")
plt.legend(["Oscillations en fonction du temps","Enveloppe d'oscillation"])
```

#### 4) Tangente à une courbe en un point donné - Application en physique
Considérons une fonction $f(x)$. 
Le coefficient directeur de la tangente au graphe de $f$ au point $x_0$ est $f'(x_0)$.

L'équation de la tangente au point $(x_0,f(x_0))$ est donc :

$y \ = (x-x_0) \ f'(x_0) \ + f(x_0) $

Tracer la fonction $z = \ f(x)$ = $- 3 \ x^2 \ + 20 \ x +5 $, pour $x$ variant de 0 à 10.
Tracer ensuite les tangentes aux points $x$=0, $x$=3.333, $x$=6.905 (on définira au préalable la fonction dérivée).
Que représentent les coefficients directeurs de ces tangentes dans le cas où $x$ représente le temps et $f(x)$ représente l'altitude d'un petit objet mobile.


```python
def f(x):
    return -3*x**2+20*x+5

def f_prime(x):
    return -6*x+20

x=np.linspace(0,10,1000)
x1=0 ; x2=3.333 ; x3=6.905

def tangente(x0):
    return (x-x0)*f_prime(x0)+f(x0)


plt.plot(x,f(x),":m")
plt.plot(x,tangente(x1),"-b")
plt.plot(x,tangente(x2),"-g")
plt.plot(x,tangente(x3),"-y")
plt.title("fonction f(x)=-3*x^2+20*x+5")
plt.xlabel("x") ; plt.ylabel("f(x)")
plt.ylim(-110,100)
```
